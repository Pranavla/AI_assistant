# **Importing installed libraries**

```python
!pip install numba==0.55.1
import whisper 
import gradio as gr
import openai
from TTS.api import TTS
```

# **Creating the model**

```python
model_name = TTS.list_models()[9]
tts = TTS(model_name)
```

```python
tts.tts_to_file(text = "Hi,how are you", file_path = "output.wav")
```

# **Testing the model**

```python
from IPython.display import Audio, display

display(Audio("output.wav", autoplay = True))
```

```python
model = whisper.load_model("medium")
```

# **Importing open api key**

```python
openai.api_key = "s"
```

```python
def speak(user_voice):
  messages = [{"role" : "system", "content" : "You are a lind helpful assistant"}]
  user_message = model.transcribe(user_voice)["text"]
  messages.append({"role" : "user" , "content" : user_message})
  print(messages)
  chat = openai.ChatCompletion.create(model = "gpt-3.5-turbo" , messages = messages)
  reply = chat.choices[0].message.content
  messages.append({"role" : "assistant" , "content" :reply})
  tts.tts_to_file(text = reply,file_path = "output.wav")
  return(reply, "output.wav")
```

## **Model excecution**

```python
text_reply = gr.Textbox(label = "chatGPT")
voice_reply = gr.Audio("output.wav",label = "Voice")

gr.Interface(title = "AI assistant",
             fn = speak,
             inputs = [gr.inputs.Audio(source = "microphone" , 
                                       type = "filepath")],
         outputs = [text_reply, voice_reply
                    ], live = True).launch(debug = True,share=True)
```
